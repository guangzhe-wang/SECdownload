```python
import urllib.request
import io
import requests
import time
i = 0
folder_name = '/content/data1/'
os.makedirs(folder_name, exist_ok=True)
with open('filing_urls.txt', 'r') as f:
    urls = f.readlines()
for url in urls:
    url = url.strip()  # 去除 url 前后的空格和换行符
    filename = url.split('/')[-1]  # 获取文件名，最后一个斜线后的部分
    print(filename)
    i = i + 1
    send_headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36",
        "Connection": "keep-alive",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8",
        "Accept-Language": "zh-CN,zh;q=0.8"}
    response = requests.get(url, headers=send_headers)
    bytes_io = io.BytesIO(response.content)
    with open(folder_name + filename, 'wb') as f:
      f.write(response.content)
    time.sleep(2)
print(i)
```
